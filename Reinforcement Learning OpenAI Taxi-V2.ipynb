## Reinforcement Learning | OpenAI | Taxi-v2 | Q-Learning

- Numpy for our Qtable
- OpenAI Gym for our Taxi Environment
- Random to generate random numbers

```python
import numpy as np
import gym
import random
```

## Step 1: Create Environment ðŸŽ®
   - Here we'll create the Taxi environment.
   - OpenAI Gym is a library composed of many environments that we can use to train our agents.

```python
env = gym.make('Taxi-v2')
env.render()
```

## Step 2: Create the Q-table and initialize it ðŸ—„ï¸
   -  create our Q-table, to know how much rows (states) and columns (actions). calculate the action_size and the state_size
   - OpenAI Gym provides us a way to do that: `env.action_space.n` and `env.observation_space.n`

```python
action_size = env.action_space.n
print('action_size:{}'.format(action_size))

state_size = env.observation_space.n
print('state_size:{}'.format(state_size))
```

```python
qtable = np.zeros((state_size, action_size))
qtable.shape
```

## Step 3: The hyperparameters âš™ï¸
   - specify the hyperparameters

```python
episodes = 50000
test_episodes = 100
steps = 99

alpha = 0.7
gamma = 0.618

epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.01
```

## Step 4: The Q learning algorithm ðŸ•¹ï¸ðŸš•
  - Now we implement the Q learning algorithm:

```python
for e in range(episodes):
    done = False
    s = 0
    state = env.reset()
    
    for s in range(steps):
        exp_exp = random.uniform(0,1)      
        if exp_exp > epsilon:
            action = np.argmax(qtable[state,:])
        else:
            action = env.action_space.sample()
        new_state, reward, done, info = env.step(action)
        # Update Q(s,a):= Q(s,a) + lr [R(s,a) + gamma * max Q(s',a') - Q(s,a)]
        qtable[state, action] = qtable[state, action] + alpha * (reward + gamma * np.max(qtable[new_state,:]) - qtable[state, action])
        
        state = new_state
        if done:
            break
    epsilon = min_epsilon + (max_epsilon - min_epsilon)* np.exp(-e*decay_rate)      
    
```

## Step 5: Use trained Q-table and play Taxi ! ðŸš–
   - After 50000 episodes, the Q-table can be used as a \"cheatsheet\" to play Taxi.
   - By running the below cell we can see our agent playing Taxi.

```python
env.reset()
rewards =[]

for e in range(test_episodes):
    done = False
    state = env.reset()
    s = 0
    for s in range(steps):
        env.render()
        action = np.argmax(qtable[state,:])
        new_state, reward, done, info = env.step(action)
        reward += reward
        state = new_state
        if done:
            rewards.append(reward)
            break
env.close()    
print('the score is: {}'.format(sum(rewards)/test_episodes))        
            
```
